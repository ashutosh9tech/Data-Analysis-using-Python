```python
#Pyplot API (state-based)
I = np.linspace(0, 10, 100)   # Current in Amps
R = 5                         # Ohms
V = R * I                     # V = IR

plt.plot(I, V, label="V = IR", color="red")
plt.xlabel("Current (A)")
plt.ylabel("Voltage (V)")
plt.title("Ohm's Law")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.subplot(221)  # Top-left
plt.plot([1,2,3], [1,4,9], marker="o", color="red")
plt.grid()
plt.title("Plot 1")

plt.subplot(2,2,2)  # Top-right
plt.bar([1,2,3], [1,2,3], color=["purple", "brown", "magenta"])
plt.title("Plot 2")

plt.subplot(2,2,(3,4))  # Bottom-right
plt.plot([1,2,3], [2,3,5],marker="o")
plt.title("Plot 3")

plt.tight_layout()
plt.show()
```

```python
#Object-oriented API

fig, axs = plt.subplots(2, 3, figsize=(12,7))


axs[0,0].plot(df['age'], df['salary'], color="red", marker="o")
axs[0,0].grid()
axs[0,0].set_title("Line plot")
axs[0,0].set_xlabel("Age")
axs[0,0].set_ylabel("Salary")

axs[0,1].hist(df['salary'], bins=4, color="skyblue")
axs[0,1].set_title("Histogram")
axs[0,1].set_xlabel("Salary")
axs[0,1].set_ylabel("Frequency")

axs[0,2].pie(count, labels=count.index, autopct="%1.0f")
axs[0,2].set_title("Pie Chart")
axs[0,2].set_xlabel("Department")

axs[1,0].bar(["HR", "IT", "Finance"], [hr, it, fin], color=["orange", "green", "blue"], label=["HR", "IT", "F"])
axs[1,0].set_title("Bar Chart")
axs[1,0].set_xlabel("Department")
axs[1,0].set_ylabel("Average of salary")
axs[1,0].legend()

axs[1,1].scatter(df['age'], df['salary'], c="red")
axs[1,1].set_title("Scatter Plot")
axs[1,1].set_xlabel("Age")
axs[1,1].set_ylabel("Salary")
axs[1,1].grid()

axs[1,2].boxplot(df['salary'])
axs[1,2].set_title("Box Plot")
axs[1,2].set_xlabel("Salary")


plt.tight_layout()
plt.savefig("OO.png")
```

# Seaborn

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
penguins.dropna(inplace=True) # drop all missing values (NaN or None)
penguins
```

```python
# scatter plot
sns.scatterplot(data=penguins, x="species", y="bill_length_mm", hue="island", style="sex", alpha=0.9)
```

```python
# strip plot
sns.stripplot(data=penguins, x="species", y="bill_length_mm", hue="island",dodge=True, jitter=True)
```

```python
# swarm plot
sns.stripplot(data=penguins, x="species", y="bill_length_mm", hue="island", dodge=True, jitter=True)
```

```python
# Histogram
sns.histplot(penguins, x="body_mass_g", hue="sex", multiple="stack")
```

```python
# regression
sns.regplot(penguins, x="body_mass_g", y="flipper_length_mm")
```

```python
# line plot
sns.set_style("whitegrid")
sns.lineplot(penguins, x= "body_mass_g", y="flipper_length_mm", hue="island", style="sex")
```

```python
# joint plot
sns.set_style("whitegrid")
sns.jointplot(penguins, x="body_mass_g", y="flipper_length_mm",  kind="reg")
```

```python
# bar plot
sns.barplot(penguins, x= "species", y="body_mass_g", hue="sex", estimator="sum")
```

```python
penguins.groupby(['sex','species'])['body_mass_g'].sum()
```

```python
# box plot
sns.boxplot(penguins, x= "species", y="body_mass_g", hue="sex")
```

```python
# violin plot
sns.violinplot(penguins, x= "species", y="body_mass_g", hue="sex")
```

```python
sns.violinplot(penguins, x= "species", y="body_mass_g", palette="Set3")
sns.swarmplot(penguins, x= "species", y="body_mass_g", hue="island", size=4)
```

```python
# KDE (Kernal Density Estimate)
sns.kdeplot(penguins, x="body_mass_g", hue="sex", fill=True)
```

```python
penguins[['bill_length_mm', 'bill_depth_mm',	'flipper_length_mm', 'body_mass_g']].corr()
```

```python
# heatmap
sns.heatmap(penguins[['bill_length_mm', 'bill_depth_mm',	'flipper_length_mm', 'body_mass_g']].corr(), annot=True)
```

```python
# rugplot
sns.rugplot(penguins, x="body_mass_g", hue="species", height=0.5)
```

```python
# pairplot
sns.pairplot(penguins, hue="sex")
```

```python
# pairgrid
graph = sns.PairGrid(penguins, hue="species")
graph.map_lower(sns.scatterplot)
graph.map_upper(sns.kdeplot)
graph.map_diag(sns.histplot)
graph.add_legend()
```
